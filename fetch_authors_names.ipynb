```python
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint

import os
import dotenv
dotenv.load_dotenv()
import pandas as pd

from mybooks.library import Library
```

## https://www.ranker.com

```python
urls = [
    "https://www.ranker.com/crowdranked-list/best-fantasy-authors",
    "https://www.ranker.com/crowdranked-list/best-crime-authors",
    "https://www.ranker.com/list/best-french-authors/ranker-books",
    "https://www.ranker.com/crowdranked-list/greatest-science-fiction-authors-v1"
]
```

```python
authors_names = []
for url in urls:
    print(url)
    result = requests.get(url)
    soup = BeautifulSoup(result.text)
    for li_ele in soup.find_all('li'):
        if li_ele.has_key('class'):
            if li_ele['class'][0].startswith('gridItem_main'):
                try:
                    author = list(li_ele.children)[4].h2.text
                except:
                    author = list(li_ele.children)[3].h2.text
                if author not in authors_names:
                    print(author)
                    authors_names.append(author)
    sleep(randint(2,4))
authors_names
```

```python
library = Library(api_key=os.getenv('API_KEY'))
for search in authors_names:
    print(f"Search for {search}")
    library.fetch(author=search)
library.save_library()
print(library.library.shape)
```
